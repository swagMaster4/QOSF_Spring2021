# **Quantum Circuit Simulator**

Just as a single qubit has two possible states, 

${{\begin{bmatrix}1\\0\end{bmatrix}}}$ & ${{\begin{bmatrix}0\\1\end{bmatrix}}}$ 

the possible states for two qubits are: 

${{\begin{bmatrix}00\\01\\10\\11\end{bmatrix}}}$

Generally, the states of any qubit are $2^n$ elements. Thus, we use the tensor product to represent the combined state of multiple qubits, as shown below for two qubits. 

${\displaystyle \vert ab\rangle =\vert a\rangle \otimes \vert b\rangle =v_{00}\vert 00\rangle +v_{01}\vert 01\rangle +v_{10}\vert 10\rangle +v_{11}\vert 11\rangle \rightarrow {\begin{bmatrix}v_{00}\\v_{01}\\v_{10}\\v_{11}\end{bmatrix}}}$

Similarly, as we compute the tensor product for 3 qubits, it become evident that the size of the state vector will be $2^n$ elements

Now if we wanted to apply the X gate to a multi qubit circuit, which will be 3 qubits in our examples, the dimensions for matrix multiplication do not match between the X gate and the 3 qubit state vector:



$X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$ Dimension, (2 X 2) 

Ground State Vector $= \begin{bmatrix}1\\0\\0\\0\\0\\0\\0\\0\end{bmatrix}$ Dimension, (8 X 1) 


Thus we must find a matrix operator that ensures the gate's dimension matches that of the state vector, in this case 8. The following tensor product increases the dimension to 4 X 4. 

Screen Shot 2021-02-20 at 10.35.20 AM.png

Doing the tensor product again with the identity matrix: ${X \otimes I \otimes I}$ in this order is essentially applying the X gate to the first qubit. This results in an operator matrix that matches the state vector.

For single-qubit gate $U$ in 3-qubit circuit:



gate on qubit 0: ${O=U \otimes I \otimes I}$

gate on qubit 1: ${O=I \otimes U \otimes I}$

gate on qubit 2: ${O=I \otimes I \otimes U}$


Now let's look at how this pattern extends: 


For 3-qubit circuit, CNOT gate (a two qubit gate):

acting on first and second qubit, operator is ${O=CNOT \otimes I}$

acting on second and third qubit, operator is ${O=I \otimes CNOT}$


Example for single-qubit gate $U$ in 4-qubit circuit:

gate on qubit 0: ${O=U \otimes I \otimes I \otimes I }$

gate on qubit 1: ${O=I \otimes U \otimes I \otimes I}$

gate on qubit 2: ${O=I \otimes I \otimes U \otimes I}$

gate on qubit 3: ${O=I \otimes I \otimes I \otimes U}$


So then for 4-qubit circuit, CNOT gate:

acting on first and second qubit, operator is ${O=CNOT \otimes I \otimes I}$

acting on second and third qubit, operator is ${O=I \otimes CNOT \otimes I}$

acting on second and third qubit, operator is ${O=I \otimes I \otimes CNOT }$

The above patterns take care of:

Case 1: any single qubit unitary gate operation

Case 2: CNOT gate if the target and control qubit are adjacent
